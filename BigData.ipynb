# **Rad sa upitima**

```python
import pandas as pd

def agg_to_df(pipeline):
    return pd.DataFrame(list(collection.aggregate(pipeline, allowDiskUse=True)))
# brojevi su se zavrsavali sa e+, pa postavljam sve na dve decimale
pd.set_option("display.float_format", "{:,.2f}".format)
```

**Istraivako pitanje 1**

---

*Kako su se kretale cene nekretnina tokom perioda 20012022 i da li postoje jasni vremenski trendovi ili anomalije?*

```python
# Trend cena po godinama
q1_1_trend_year = [
  {"$match": {"Year": {"$ne": None}, "Sale Amount": {"$ne": None}}},
  {"$group": {
      "_id": "$Year",
      "avg_sale": {"$avg": "$Sale Amount"},
      "min_sale": {"$min": "$Sale Amount"},
      "max_sale": {"$max": "$Sale Amount"},
      "count": {"$sum": 1}
  }},
  {"$sort": {"_id": 1}}
]
df_q1_1 = agg_to_df(q1_1_trend_year)

df_q1_1 = agg_to_df(q1_1_trend_year)
df_q1_1
```

```python
# Ukupna vrednost trita po godini
q1_2_total_value = [
  {"$match": {"Year": {"$ne": None}, "Sale Amount": {"$ne": None}}},
  {"$group": {
      "_id": "$Year",
      "total_sales_value": {"$sum": "$Sale Amount"},
      "count": {"$sum": 1}
  }},
  {"$sort": {"_id": 1}}
]
df_q1_2 = agg_to_df(q1_2_total_value)

df_q1_2 = agg_to_df(q1_2_total_value)
df_q1_2.head(10)

```

```python
# Sezonalnost prodaje (meseci)
q1_3_seasonality = [
  {"$match": {"Date Recorded": {"$ne": None}, "Sale Amount": {"$ne": None}}},
  {"$addFields": {"month": {"$month": "$Date Recorded"}}},
  {"$group": {
      "_id": "$month",
      "count": {"$sum": 1},
      "avg_sale": {"$avg": "$Sale Amount"}
  }},
  {"$sort": {"_id": 1}}
]
df_q1_3 = agg_to_df(q1_3_seasonality)

df_q1_3 = agg_to_df(q1_3_seasonality)
df_q1_3.head(10)

```

```python
# Anomalije (najskuplje prodaje)
q1_4_outliers = [
  {"$match": {"Sale Amount": {"$ne": None}}},
  {"$sort": {"Sale Amount": -1}},
  {"$project": {
      "_id": 0,
      "Town": 1,
      "Address": 1,
      "Year": 1,
      "Sale Amount": 1,
      "Property Type": 1
  }},
  {"$limit": 10}
]
df_q1_4 = agg_to_df(q1_4_outliers)

df_q1_4 = agg_to_df(q1_4_outliers)
df_q1_4.head(10)

```

**Istraivako pitanje 2**

---
*U kojoj meri lokacija utie na prodajnu cenu nekretnina?*


```python
# Najskuplji gradovi (prosek)
q2_1_top_price_towns = [
  {"$match": {"Town": {"$ne": None}, "Sale Amount": {"$ne": None}}},
  {"$group": {"_id": "$Town", "avg_sale": {"$avg": "$Sale Amount"}, "count": {"$sum": 1}}},
  {"$match": {"count": {"$gte": 200}}},
  {"$sort": {"avg_sale": -1}},
  {"$limit": 10}
]
df_q2_1 = agg_to_df(q2_1_top_price_towns)

df_q2_1 = agg_to_df(q2_1_top_price_towns)
df_q2_1.head(10)

```

```python
# Najaktivniji gradovi (broj prodaja)
q2_2_most_active_towns = [
  {"$match": {"Town": {"$ne": None}}},
  {"$group": {"_id": "$Town", "count": {"$sum": 1}}},
  {"$sort": {"count": -1}},
  {"$limit": 10}
]
df_q2_2 = agg_to_df(q2_2_most_active_towns)

df_q2_2 = agg_to_df(q2_2_most_active_towns)
df_q2_2.head(10)

```

```python
# Price_Diff po gradu
q2_3_price_diff_town = [
  {"$match": {"Town": {"$ne": None}, "Price_Diff": {"$ne": None}}},
  {"$group": {"_id": "$Town", "avg_diff": {"$avg": "$Price_Diff"}, "count": {"$sum": 1}}},
  {"$match": {"count": {"$gte": 200}}},
  {"$sort": {"avg_diff": -1}},
  {"$limit": 10}
]
df_q2_3 = agg_to_df(q2_3_price_diff_town)

df_q2_3 = agg_to_df(q2_3_price_diff_town)
df_q2_3.head(10)

```

**Istraivako pitanje 3**

---
*Kako karakteristike nekretnine utiu na prodajnu cenu?*


```python
# Cena po Property Type
q3_1_property_type = [
  {"$match": {"Property Type": {"$ne": None}, "Sale Amount": {"$ne": None}}},
  {"$group": {"_id": "$Property Type", "avg_sale": {"$avg": "$Sale Amount"}, "count": {"$sum": 1}}},
  {"$sort": {"avg_sale": -1}}
]
df_q3_1 = agg_to_df(q3_1_property_type)

df_q3_1 = agg_to_df(q3_1_property_type)
df_q3_1.head(10)

```

```python
# Cena po Residential Type
q3_2_residential_type = [
  {"$match": {"Residential Type": {"$ne": None}, "Sale Amount": {"$ne": None}}},
  {"$group": {"_id": "$Residential Type", "avg_sale": {"$avg": "$Sale Amount"}, "count": {"$sum": 1}}},
  {"$sort": {"count": -1}}
]
df_q3_2 = agg_to_df(q3_2_residential_type)

df_q3_2 = agg_to_df(q3_2_residential_type)
df_q3_2.head(10)

```

```python
# Sales Ratio po tipu nekretnine
q3_3_sales_ratio = [
  {"$match": {"Property Type": {"$ne": None}, "Sales Ratio": {"$ne": None}}},
  {"$group": {
      "_id": "$Property Type",
      "avg_ratio": {"$avg": "$Sales Ratio"},
      "count": {"$sum": 1}
  }},
  {"$sort": {"avg_ratio": -1}}
]
df_q3_3 = agg_to_df(q3_3_sales_ratio)

df_q3_3 = agg_to_df(q3_3_sales_ratio)
df_q3_3.head(10)

```

```python
# Cenovne klase (segmentacija)
q3_4_price_segments = [
  {"$match": {"Sale Amount": {"$ne": None}}},
  {"$bucketAuto": {
      "groupBy": "$Sale Amount",
      "buckets": 5,
      "output": {
          "count": {"$sum": 1},
          "avg_assessed": {"$avg": "$Assessed Value"}
      }
  }}
]
df_q3_4 = agg_to_df(q3_4_price_segments)

df_q3_4 = agg_to_df(q3_4_price_segments)
df_q3_4.head(10)

```

```python
# Godine sa najveim rastom/padom
df_q6_1 = df_q1_1[["_id", "avg_sale"]].rename(columns={"_id": "Year"}).sort_values("Year")
df_q6_1["prev_avg"] = df_q6_1["avg_sale"].shift(1)
df_q6_1["yoy_change"] = df_q6_1["avg_sale"] - df_q6_1["prev_avg"]

df_q6_1.head(10)

```

```python
# Kombinacija lokacija + tip
top_towns = df_q2_2["_id"].head(5).tolist()

q6_2_town_property = [
  {"$match": {"Town": {"$in": top_towns}, "Property Type": {"$ne": None}}},
  {"$group": {
      "_id": {"Town": "$Town", "Property Type": "$Property Type"},
      "avg_sale": {"$avg": "$Sale Amount"},
      "count": {"$sum": 1}
  }},
  {"$sort": {"_id.Town": 1, "avg_sale": -1}}
]
df_q6_2 = agg_to_df(q6_2_town_property)

df_q6_2 = agg_to_df(q6_2_town_property)
df_q6_2.head(10)

```
